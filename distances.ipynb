```python
import numpy as np
import matplotlib.pyplot as plt
import math
import torch
import torch.distributions as ds
import scipy.stats
```

```python
P = ds.Uniform(0,1)
Q1 = ds.Uniform(0.4,1)
```

```python
def Q2():
    U1 = ds.Uniform(0,0.77815)
    U2 = ds.Uniform(0.77815, 1)
    return 0.285 * U1.sample() + 3.479 * U2.sample()
```

```python
def JS(p, q):
    p /= p.sum()
    q /= q.sum()
    m = (p + q) / 2
    return (scipy.stats.entropy(p, m) + scipy.stats.entropy(q, m)) / 2
```

```python
P = np.ones(100000)
Q1 = np.concatenate((np.zeros(39999), np.ones(60001)))
Q2 = np.concatenate((0.285*np.ones(77815), np.zeros(22185))) + np.concatenate((np.zeros(77815), 3.479 * np.ones(22185)))
```

```python
print(JS(P**3,Q1**3))
```

```python
N = 100000
P_hat = np.zeros(N)
Q1_hat = np.zeros(N)
Q2_hat = np.zeros(N)
for i in range(100000):
    psample = P.sample()
    P_hat[int(psample*N)] += 1
    q1sample = Q1.sample()
    Q1_hat[int(q1sample*N)] += 1
    q2sample = Q2()
    Q2_hat[int(q2sample*N)] += 1
```

```python
print(JS(P_hat,Q1_hat))
```
